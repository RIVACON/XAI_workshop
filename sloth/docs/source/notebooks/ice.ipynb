# Individual Conditional Expectations (ICE)

```python
import numpy as np
import sys
import seaborn as sn
import config
import matplotlib.pyplot as plt
import sloth


import logging
logging.basicConfig()
logger = logging.getLogger('sloth')
logger.setLevel(level=logging.DEBUG)

sn.set_style("whitegrid")
#sn.set_theme()

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
# get a sample validation task from a synthetic credit default model
validation_task = sloth.datasets.credit_default.get(0,0)
```

```python
ice = sloth.ICE(validation_task, max_samples=100)
ice_explanation = ice.explain(x=None)
```

```python
plt.figure(figsize=(12,9))
ice_explanation.plot(n_subplot_cols=2, clustering=False)
```
